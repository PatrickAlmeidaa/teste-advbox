```python
import pandas as pd
```

```python
df = pd.read_csv('dados_advbox.csv')
```

```python
df.head()
```

```python
# Limpeza dos números de telefone
df['Telefone'] = df['Telefone'].str.replace('Celular comercial:', '')
df['Telefone'] = df['Telefone'].str.replace('Celular pessoal:', '')
df['Telefone'] = df['Telefone'].str.replace('Telefone comercial:', '')
df['Telefone'] = df['Telefone'].str.replace('Telefone residencial:', '')
df['Telefone'] = df['Telefone'].str.replace('Telefone:', '')
df['Telefone'] = df['Telefone'].str.replace('Whatsapp:', '')
df['Telefone'] = df['Telefone'].str.replace('(', '')
df['Telefone'] = df['Telefone'].str.replace(')', '')
df['Telefone'] = df['Telefone'].str.replace(' ', '')
df['Telefone'] = df['Telefone'].apply(lambda x: x[:11])  # Pega os primeiros 11 dígitos
df['Telefone'] = df['Telefone'].apply(lambda x: x if len(x) == 11 else '0' + x)  # Adiciona 0 no início se faltar dígitos
```

```python
df.head(33)
```

```python
# Divisão do endereço
endereco = df['Endereço'].str.split(' - ', expand=True)
df['Endereço'] = endereco[0]
df['Bairro'] = endereco[1]
df['Cidade'] = endereco[2]
df['Estado'] = endereco[3].str.split(' ', expand=True)[0]  # Assume que a sigla do estado é o primeiro elemento
df['CEP'] = endereco[4].str.replace(r'\D+', '')  # Remove todos os caracteres não numéricos do CEP
```

```python
df.head(34)
```

```python
df['PAIS'] = ''
df['ANOTAÇÕES GERAIS'] = ''
```

```python
df = df[['Nome/Empresa', 'Telefone', 'PAIS', 'Estado', 'Cidade', 'Bairro', 'Endereço', 'CEP', 'ANOTAÇÕES GERAIS']]
```

```python
# Salvar a nova tabela
df.to_excel('modelo_base_padrao.xlsx', index=False)
```
